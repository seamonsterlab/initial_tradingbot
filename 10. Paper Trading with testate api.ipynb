```python
def get_history(symbol, interval, start, end = None):
    bars = client.get_historical_klines(symbol = symbol, interval = interval, start_str = start, end_str = end, limit = 1000)
    df = pd.DataFrame(bars)
    df["Date"] = pd.to_datetime(df.iloc[:,0], unit = "ms")
    df.columns = ["Open Time", "Open", "High", "Low", "Close", "Volume", "Clos time", "Quote Asset Volume", "Number of Trades", "Taker Buy Base Asset Volume", "Taker Buy Quote Asset Volume", "Ignore", "Date"]
    df = df[["Date", "Open", "High", "Low", "Close", "Volume"]].copy()
    df.set_index("Date", inplace =True)
    for column in df.columns:
        df[column] = pd.to_numeric(df[column], errors = "coerce")
    return df
```

```python
df = get_history(symbol = "BTCUSDT", interval = "1d", start = timestamp)
df
```

```python
from binance import ThreadedWebsocketManager
```

```python
def stream_data(msg):
    time = pd.to_datetime(msg["E"], unit = "ms")
    price = msg["c"]
    print("Time: {} | Price: {}".format(time, price))
```

```python
twm = ThreadedWebsocketManager()
twm.start()
```

```python
twm.start_symbol_miniticker_socket(callback = stream_data, symbol = "BTCUSDT")
```
